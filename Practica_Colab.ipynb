```python
!pip install biopython
from Bio import pairwise2
from Bio.Seq import Seq
from Bio.pairwise2 import format_alignment
```

```python
#Crear análisis de secuencias
seq1 = Seq("AGTTTGATCA")
seq2 = Seq("ATTTGATCCT")
```

```python
#Encontrar similitudes
alignments = pairwise2.align.globalxx(seq1, seq2)
```

```python
#Resultados
for match in alignments:
    print(match)
```

```python
#Calcular el porcentaje
secuencia_adn = input("Ingresa una secuencia de ADN: ")
secuencia_adn = secuencia_adn.upper() # Convert the input to uppercase
total_nucleotidos = len(secuencia_adn) # Calculate the length of the sequence
num_g = secuencia_adn.count('G')
num_c = secuencia_adn.count('C')
porcentaje_gc = (num_g + num_c) / total_nucleotidos * 100
print(f"Porcentaje de GC: {porcentaje_gc:.2f}%")
```
